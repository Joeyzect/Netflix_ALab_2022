```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movie_val_auc = [0.83188, 0.82017, 0.82577, 0.60497]
movie_test_auc = [0.85026, 0.83014, 0.83451, 0.64252]

tvshow_val_auc = [0.800, 0.806]
tvshow_test_auc = [0.857, 0.876]

movie_methods = ['No Storyline','Bag of Words','TF-IDF','BERT']
tvshow_methods = ['No Storyline','Bag of Words']

# generate multiple bar chart for movie
plt.figure(figsize=(10, 6))
plt.bar(movie_methods, movie_val_auc, color='black', width=0.2, label='Validation AUC')
plt.bar([i + 0.2 for i in range(len(movie_methods))], movie_test_auc, color='#e50914', width=0.2, label='Test AUC')
# center the x-axis labels
plt.xticks([i + 0.1 for i in range(len(movie_methods))], movie_methods)

# print the values on top of the bars to 3 decimal places
for i in range(len(movie_val_auc)):
    plt.text(x=i - 0.10, y=movie_val_auc[i] + 0.01, s=round(movie_val_auc[i], 3), size=10)
    plt.text(x=i + 0.13, y=movie_test_auc[i] + 0.01, s=round(movie_test_auc[i], 3), size=10)

plt.xlabel('Methods')
plt.ylabel('AUC')
plt.title('Movie Validation and Test AUC')
plt.legend()
plt.show()
```

```python
# generate multiple bar chart for tvshow
plt.figure(figsize=(10, 6))
plt.bar(tvshow_methods, tvshow_val_auc, color='black', width=0.2, label='Validation AUC')
plt.bar([i + 0.2 for i in range(len(tvshow_methods))], tvshow_test_auc, color='#e50914', width=0.2, label='Test AUC')
# center the x-axis labels
plt.xticks([i + 0.1 for i in range(len(tvshow_methods))], tvshow_methods)

# print the values on top of the bars to 3 decimal places
for i in range(len(tvshow_val_auc)):
    plt.text(x=i - 0.05, y=tvshow_val_auc[i] + 0.01, s=round(tvshow_val_auc[i], 3), size=10)
    plt.text(x=i + 0.15, y=tvshow_test_auc[i] + 0.01, s=round(tvshow_test_auc[i], 3), size=10)

plt.xlabel('Methods')
plt.ylabel('AUC')
plt.title('TV Show Validation and Test AUC')
plt.legend()
plt.show()
```
