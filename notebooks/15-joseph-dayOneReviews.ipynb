```python
params = {
    'ref_': 'undefined',
    'paginationKey': ''
}

link = 'https://www.imdb.com/title/tt13433812/reviews/_ajax'
```

```python
#get all reviews
reviews = []
while True:
    response = requests.get(link, params=params)
    soup = BeautifulSoup(response.text, 'html.parser')
    reviews.extend(soup.find_all('div', class_='imdb-user-review'))
    if soup.find('div', class_='load-more-data') is None:
        break
    params['paginationKey'] = soup.find('div', class_='load-more-data')['data-key']
```

```python
with requests.Session() as s:
    s.headers['User-Agent'] = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36'
    res = s.get(url)

    while True:
        soup = BeautifulSoup(res.text,"lxml")
        for item in soup.select(".review-container"):
            reviewer_name = item.select_one("span.display-name-link > a").get_text(strip=True)
            print(reviewer_name)

        try:
            pagination_key = soup.select_one(".load-more-data[data-key]").get("data-key")
        except AttributeError:
            break
        params['paginationKey'] = pagination_key
        res = s.get(link,params=params)
```

```python
# get reviews with min date
reviews = soup.find_all('div', class_='lister-item-content')
# review = reviews[0]
# review.find('a', class_='title').text.strip()
# review.find('span', class_='display-name-link').text.strip()
# review.find('span', class_='review-date').text.strip()
```

```python
#get all reviews
all_reviews = []
for review in reviews:
    title = review.find('a', class_='title').text.strip()
    name = review.find('span', class_='display-name-link').text.strip()
    date = review.find('span', class_='review-date').text.strip()
    all_reviews.append([title, name, date])
```
