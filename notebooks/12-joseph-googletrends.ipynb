```python
import time
import random
import pickle
```

```python
mylist601 = []
for i in range(601, len(orig_netflix_titles)):
    
    mylist601.append(t.new_query(orig_netflix_titles['originalTitle'].iloc[i]))
    time.sleep(10+random.random()*5)
    #pickle the list every 100 queries
    if i % 100 == 0:
        with open(f'../data/interim/mylist{i}.pkl', 'wb') as f:
            pickle.dump(mylist601, f)
```
